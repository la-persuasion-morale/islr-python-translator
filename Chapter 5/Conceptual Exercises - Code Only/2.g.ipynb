```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def prob(n):
    return(1 - pow((1 - 1/n),n))
```

```python
x = np.array([])
for i in range(1, 1000):
    x = np.append(x, prob(i))
```

```python
plt.xkcd()
plt.figure(figsize = (25, 10))
plt.plot(x, color = 'green')
plt.xlabel("x")
plt.ylabel("pr(x)")
plt.title("Plot for bootstrap")
```

**We see that the plot quickly reaches an asymptote of 0.63.**
